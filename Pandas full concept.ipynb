## pandas.Series.str.split

```python
s = pd.Series(["this is a regular sentence",
               "https://docs.python.org/3/tutorial/index.html",
               np.nan])
```

```python
s
```

```python
s.str.split()
```

```python
s.str.rsplit()
```

```python
s.str.split(n=2)
```

```python
s.str.rsplit(n=2) ##The outputs of split and rsplit are different.
```

When using expand=True, the split elements will expand out into separate columns.
If NaN is present, it is propagated throughout the columns during the split.

```python
s.str.split(expand=True)
```

### pandas.DataFrame.replace

```python
s = pd.Series([0, 1, 2, 3, 4])
s.replace(0, 5)
```

```python
df = pd.DataFrame({'A': [0, 1, 2, 3, 4],
                   'B': [5, 6, 7, 8, 9],
                   'C': ['a', 'b', 'c', 'd', 'e']})
df.replace(0, 5)
```

```python
df.replace(2,6)
```
